```python
from openai import OpenAI
import requests
from bs4 import BeautifulSoup
import json

# API Setup
api_key = "ddc-a4f-49e9ee762ff445bcb1146ffb238d75b3"
api_base_url = "https://api.a4f.co/v1"
client = OpenAI(api_key=api_key, base_url=api_base_url)

def extract_links(company_url):
    response = requests.get(company_url, timeout=10)
    soup = BeautifulSoup(response.text, "html.parser")
    base_url = company_url.rstrip("/")  # Ensure correct base URL

    # Collect all links from the page
    links = set()
    for a in soup.find_all("a",href = True):
      links.add(a["href"])
    print(links)
    structured_links = []
    for link in links:
      if link.startswith("/"):
        structured_links.append(base_url + link)
      else:
        structured_links.append(link)
    print(structured_links)


def generate_brochure(company_name, company_url):
    # Step 1: Extract relevant links
    relevant_links = extract_links(company_url)

    # Step 2: Use GPT-4o-Mini to generate structured JSON
    prompt = {
    "role": "system",
    "content": """Extract details from all company web pages and structure them into JSON format.
    Example Response:
    {
        "company_name": "Sample Corp",
        "website": "https://samplecorp.com",
        "pages": {
            "/about": "Sample Corp is a leading provider of AI solutions.",
            "/careers": "Sample Corp is hiring AI engineers.",
            "/investors": "Investing in Sample Corp provides long-term tech growth.",
            "/products": "Sample Corp offers AI-powered tools for businesses.",
            "/blog": "Read about industry trends and innovations.",
            "/news": "Latest updates and announcements from Sample Corp.",
            "/events": "Upcoming conferences and webinars."
        }
    }
    """
}



    messages = [
        prompt,
        {"role": "user", "content": f"Company: {company_name}, Website: {company_url}, Relevant Links: {relevant_links}"}
    ]

    completion = client.chat.completions.create(
        model="provider-4/gpt-4o-mini",
        messages=messages
    )

    print("RAW RESPONSE:\n", completion.choices[0].message.content)

# Example Usage
company_name = "Accenture"
company_url = "https://www.accenture.com/us-en"

brochure_data = generate_brochure(company_name, company_url)

```
